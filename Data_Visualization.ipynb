```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = '/content/final_forecasted_dataset.xlsx'

# Load the relevant data (G4261:H5599) from the Excel file
df = pd.read_excel(file_path, sheet_name='data', usecols="G:H", skiprows=4260, nrows=1339)

# Renaming columns for clarity
df.columns = ['Packs Forecast', 'Value Forecast']

# Plotting the data
plt.figure(figsize=(12, 6))

# Plot for Packs Forecast
plt.subplot(2, 1, 1)
plt.plot(df['Packs Forecast'], label='Packs Forecast')
plt.title('Packs Forecast (G4261:G5599)')
plt.xlabel('Index')
plt.ylabel('Packs')
plt.legend()

# Plot for Value Forecast
plt.subplot(2, 1, 2)
plt.plot(df['Value Forecast'], label='Value Forecast', color='orange')
plt.title('Value Forecast (H4261:H5599)')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```
